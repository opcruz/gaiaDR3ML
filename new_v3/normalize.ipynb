```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
sampling = np.linspace(336, 1020, 343)
```

```python
def normalize(spectra):
    min1 = min(spectra)
    max1 = max(spectra)
    # Normalizar los espectros al rango de 0 a 1
    return (spectra - min1) / (max1 - min1)
```

## Normalize Planetary Nebulae (GAIA)

```python
df1 = pd.read_csv("./calibrated_data/NP.csv")
df1['flux'] = df1['flux'].apply(lambda x: np.array([float(i) for i in x.strip('[]').split()]))
df1['flux'] = df1['flux'].apply(lambda x: normalize(x))

dfluxes = pd.DataFrame(df1['flux'].tolist(), columns=range(343))
dfluxes.insert(0, 'source_id', df1['source_id'])

out_dir = './normalized'
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
fullname = os.path.join(out_dir, "NP.csv")

dfluxes.to_csv(fullname, header=True, index=False)
print(dfluxes.shape[0])
```

```python
# % matplotlib inline
plt.subplots(figsize=(10, 6), dpi=90)

# Create line plot
plt.plot(sampling, dfluxes.iloc[1324], c='blue')

# Set labels and title
plt.xlabel('Wavelength', fontsize=13)
plt.ylabel('Scaled Flux', fontsize=13)
plt.title('RG Star', fontsize=14)
# Display the plot
plt.show()
```
