```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from gaiaxpy import calibrate
```

```python
sampling = np.linspace(336, 1020, 343)
```

```python
df1 = pd.read_csv("./gaia-np-candidate/XP_CONTINUOUS_COMBINED.csv")
symbad = pd.read_csv("./symbad/NP_Candidate.csv", dtype={'Gaia DR3': 'int64'})
#
sy_calibrated_spectra, sampling = calibrate(df1, save_file=False)
#
symbad = symbad.dropna(subset=['Gaia DR3'])
```

```python
sy_calibrated_spectra
```

```python
symbad
```

```python
df_filtered = sy_calibrated_spectra.merge(symbad, left_on=['source_id'], right_on=['Gaia DR3'], how='inner',
                                          indicator=True)
```

```python
df_filtered
```

```python
eligible_candidates = [
    "IRAS 02379+5724",
    "UCAC2  46104304",
    "IRAS 05495+2620",
    "IRAS 06549-2330",
    "PN G228.1+00.8",
    "PN G239.3-02.7",
    "PN Y-C   40",
    "EGB  2",
    "IRAS 10348-6320",
    "IRAS 11555-6031",
    "IRAS 11415-6540",
    "PN G292.3+00.5",
    "2MASS J05221214-6949580",
    # "[RP2006] J052216-694127",
    # "[RP2006] J052254-694036",
    "[RP2006] J052438-690413",
    "[RP2006]  312"
]
```

```python
# df_filtered
eligible_np = df_filtered[df_filtered['MAIN_ID'].isin(eligible_candidates)]
eligible_np = eligible_np.reset_index(drop=True)
```

```python
eligible_np
```

```python
columns = 3
rows = 5
out_dir = "./plot"
fig, ax = plt.subplots(rows, columns, figsize=(10, 14), dpi=300, gridspec_kw={'hspace': 0.8, 'wspace': 0.25})

for index, row in eligible_np.iterrows():
    # Create line plot
    star_name = str(row['MAIN_ID'])
    i = (index % columns)
    j = int(index / columns)

    ax[j, i].plot(sampling, row['flux'], c='blue')
    ax[j, i].set_title(star_name, fontsize=7)
    ax[j, i].set_xlabel('Wavelength [nm]', fontsize=7)
    ax[j, i].set_ylabel('Flux [W nm^-1 m^-2]', fontsize=7)
    ax[j, i].tick_params(axis='both', labelsize=7)

plt.savefig(os.path.join(out_dir, 'combined.svg'), format="svg")
plt.show()
```
