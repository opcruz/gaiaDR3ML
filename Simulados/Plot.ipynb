```python

import pandas as pd
import seaborn as sns

pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
# % matplotlib inline
```

```python
#load the data
df_data_blue = pd.read_csv('../DATA/SIMULADOS21/WW/Cn2_1_BP.csv')
df_data_blue = df_data_blue[df_data_blue['wave'] < 1200]
df_data_red = pd.read_csv('../DATA/SIMULADOS21/WW//Cn2_1_RP.csv')
#df_data_red = df_data_red[df_data_red['wave'] < 1200]
```

```python
fig, ax = plt.subplots(figsize=(20, 8), dpi=90)
# plt.xticks(rotation=90)
#plt.figure(figsize=(20,8))
sns.lineplot(data=df_data_blue, x="wave", y="flux", errorbar=None,  color="blue")
sns.lineplot(data=df_data_red, x="wave", y="flux", errorbar=None,  color="red")
#plt.xlim(675,700) #Definir el rango donde queramos mostrar eje X
# ax.xaxis.set_major_locator(ticker.MultipleLocator(1000000))
plt.title('Wave And Flux')
```

Diagrama De Dispersion

```python
plt.scatter(df_data_blue['wave'], df_data_blue['flux'], color='blue', label='BP')
plt.scatter(df_data_red['wave'], df_data_red['flux'], color='red', label='RP')
plt.xlabel('wave')
plt.ylabel('flux')
plt.title('BP/RP')
plt.legend()
plt.show()
```

Grafico de Cajas

```python
plt.boxplot([df_data_blue['flux'], df_data_red['flux']])
plt.xticks([1, 2], ['BLUE', 'RED'])
plt.ylabel('wave')
plt.title('BP/RP')
plt.show()
```
