```python
import os
import zipfile
import pandas as pd
from astroquery.gaia import Gaia
```

## Download Planetary Nebulae (GAIA)

```python
df1 = pd.read_csv("./symbad/NP.csv")
ids = df1['Gaia DR3'].astype(str).values
```

```python
out_dir = './gaia'
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
fullname = os.path.join(out_dir, "NP_XP_CONTINUOUS_COMBINED.zip")

Gaia.load_data(
    ids=ids,
    format='csv',
    data_release='Gaia DR3',
    data_structure='COMBINED',
    retrieval_type='XP_CONTINUOUS',
    output_file=fullname,
    overwrite_output_file=True,
    verbose=True
)
```

## Download Red Giants (GAIA)

```python
df1 = pd.read_csv("./symbad/RG.csv")
ids = df1['Gaia DR3'].astype(str).values
```

```python
out_dir = './gaia'
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

index = 0
i = 0
size = len(ids)
while index < size:
    elements = ids[index:index + 5000]
    # Hacer algo con los elementos obtenidos, por ejemplo:
    fullname = os.path.join(out_dir, "RG{0}_XP_CONTINUOUS_COMBINED.zip".format(i))
    Gaia.load_data(
        ids=elements,
        format='csv',
        data_release='Gaia DR3',
        data_structure='COMBINED',
        retrieval_type='XP_CONTINUOUS',
        output_file=fullname,
        overwrite_output_file=True,
        verbose=True
    )

    i += 1
    print("finish", i)
    index += 5000
```
