```python
import pandas as pd
import joblib
```

LOAD DATA WITH PCA ALGORITHM IN A DATAFRAME

```python
PCA_SY  = joblib.load('../dimensionality_reduction/modelo_pca_SY.pkl')
PCA_SY['id'] = 1
PCA_SY
```

```python
PCA_NP  = joblib.load('../dimensionality_reduction/modelo_pca_NP.pkl')
PCA_NP['id'] = 2
PCA_NP
```

```python
PCA_RG  = joblib.load('../dimensionality_reduction/modelo_pca_RG.pkl')
PCA_RG['id'] = 3
PCA_RG
```

```python
nuevo_df = pd.concat([PCA_SY, PCA_NP,PCA_RG])
```

```python
nuevo_df
```

```python
nuevo_df.to_csv('spectra_pca_SNG.csv', index=False)
```
