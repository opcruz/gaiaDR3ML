```python
import matplotlib.pyplot as plt

varianza = {}
for i in range(1, 40):
  pca = PCA(n_components=i)
  pca.fit_transform(spectra_NP)
  varianza[i] = pca.explained_variance_ratio_.sum()

plt.figure(figsize=(30,20))
plt.plot(list(varianza.keys()),
         list(varianza.values()), '-o'),

plt.xlabel("numero de componentes")
plt.ylabel("varianza")
plt.xticks(list(varianza.keys()))
plt.show()
```
