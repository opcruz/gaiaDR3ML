```python
def plot_confusion_matrix(cm_balanced, cm_unbalanced, title, saved_file, save = False):
    class_names = ['SY', 'NP', 'RG']
    fig, axes = plt.subplots(1, 1, figsize=(14, 6))
    # plt.figure(figsize=(8, 5), dpi=240)
    # plt.xticks(fontsize=12)
    # plt.yticks(fontsize=12)
    sns.set(font_scale=1.6)  # Set the font scale globally
    font_props = {'fontsize': 18}
    ax = sns.heatmap(cm_unbalanced, annot=True, cmap='Greens', fmt='d', cbar=False, ax=axes[0])
    # ax.set_xlabel('Predicted Values', fontdict=font_props)
    # ax.set_ylabel('True Values', fontdict=font_props)
    ax.set_xticklabels(class_names, rotation=45, fontdict=font_props)
    ax.set_yticklabels(class_names, rotation=0, fontdict=font_props)
    ax.set_title('Unbalanced Data', fontdict=font_props)

    ax2 = sns.heatmap(cm_balanced, annot=True, cmap='Greens', fmt='d', cbar=False, ax=axes[1])
    # ax2.set_xlabel('Predicted Values', fontdict=font_props)
    # ax2.set_ylabel('True Values', fontdict=font_props)
    ax2.set_xticklabels(class_names, rotation=45, fontdict=font_props)
    ax2.set_yticklabels(class_names, rotation=0, fontdict=font_props)
    ax2.set_title('Balanced Data', fontdict=font_props)

    fig.suptitle('Confusion Matrix [{0}]'.format(title))
    fig.supxlabel('Predicted Values')
    fig.supylabel('True Values')

    # Adjust the spacing between subplots
    # Adjust the layout and spacing
    plt.tight_layout()
    plt.subplots_adjust(wspace=0.3, hspace=1.6)
    if save:
        plt.savefig(saved_file, format="svg")
    plt.show()
```

```python
def plot_confusion_matrix(cm, title, saved_file, save=False):
    class_names = ['L1', 'L2', 'L3']
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.set(font_scale=1.6)  # Set the font scale globally
    font_props = {'fontsize': 18}

    sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', cbar=False, ax=ax)
    ax.set_xticklabels(class_names, rotation=45, fontdict=font_props)
    ax.set_yticklabels(class_names, rotation=0, fontdict=font_props)
    # ax.set_title(title, fontdict=font_props)

    plt.xlabel('Predicciones', fontdict=font_props)
    plt.ylabel('Valores verdaderos', fontdict=font_props)

    plt.tight_layout()
    if save:
        plt.savefig(saved_file, format="svg")
    plt.show()
```

### RNA

```python
# Read the CSV file
cm_unbalanced_rna = np.genfromtxt('cm_rna_unbalanced.csv', delimiter=',').astype(int)
cm_balanced_rna = np.genfromtxt('cm_rna_balanced.csv', delimiter=',').astype(int)

plot_confusion_matrix(cm_balanced_rna, "RNA", "rna_combined.svg", True)
```

## Random Forest

```python
# Read the CSV file
cm_unbalanced_rf = np.genfromtxt('cm_rf_unbalanced.csv', delimiter=',').astype(int)
cm_balanced_rf = np.genfromtxt('cm_rf_balanced.csv', delimiter=',').astype(int)

plot_confusion_matrix(cm_balanced_rf, cm_unbalanced_rf, "Random Forest", "rf_combined.svg", True)
```
