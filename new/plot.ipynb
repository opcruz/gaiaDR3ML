```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
sampling = np.linspace(336, 1020, 343)
```

## Plot Planetary Nebulae (GAIA)

```python
df1 = pd.read_csv("./normalized/NP.csv")

out_dir = './plot/NP'
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

# Establecer el tamaño grande de la figura en pulgadas (ancho x altura)
fig, ax = plt.subplots(figsize=(12, 8), dpi=90)

for index, row in df1.iterrows():
    # Create line plot
    ax.plot(sampling, row, c='blue')

    # Set labels and title
    ax.set_title('NP Star', fontsize=14)
    ax.set_xlabel('Wavelength', fontsize=13)
    ax.set_ylabel('Scaled Flux', fontsize=13)

    filename = f'espectro_np_{index + 1}.png'
    fullname = os.path.join(out_dir, filename)

    # Guarda la imagen en disco
    plt.savefig(fullname)

    # Limpia el gráfico para el siguiente registro
    ax.clear()
```
