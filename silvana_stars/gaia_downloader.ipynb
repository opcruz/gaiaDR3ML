```python
import os
import zipfile
import pandas as pd
from astroquery.gaia import Gaia
```

## Download Stars

```python
df1 = pd.read_csv("./ids.csv", header=None)
ids = df1[0].astype(str).values
```

```python
ids
```

```python
out_dir = './gaia'
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

index = 0
i = 0
size = len(ids)
while index < size:
    elements = ids[index:index + 5000]
    # Hacer algo con los elementos obtenidos, por ejemplo:
    fullname = os.path.join(out_dir, "STARS_{0}_XP_CONTINUOUS_COMBINED.zip".format(i))
    Gaia.load_data(
        ids=elements,
        format='csv',
        data_release='Gaia DR3',
        data_structure='COMBINED',
        retrieval_type='XP_CONTINUOUS',
        output_file=fullname,
        overwrite_output_file=True,
        verbose=True
    )

    i += 1
    print("finish", i)
    index += 5000
```

```python
def build_query(ids):
    return """SELECT *
FROM gaiadr3.gaia_source s
LEFT JOIN gaiadr3.astrophysical_parameters p ON s.source_id = p.source_id
WHERE s.source_id IN ({}) order by s.source_id
""".format(', '.join(f"'{id}'" for id in ids))
```

```python
# ids = ids[0:4500]
```

```python
out_dir = './gaia'
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

index = 0
i = 0
size = len(ids)
out_list = []
while index < size:
    elements = ids[index:index + 2000]
    job = Gaia.launch_job(query=build_query(elements), verbose=False)
    result = job.get_results()
    out_list.append(result.to_pandas())

    i += 1
    index += 2000
    print("completed: {0} %".format(round(index * 100 / size, 2)))

out_name = 'Stars.csv'
out_dir = './gaia'
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
fullname = os.path.join(out_dir, out_name)

combined_pd = pd.concat(out_list)
combined_pd.to_csv(fullname, header=True, index=False)
```
